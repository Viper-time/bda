```python
from pyspark import SparkContext, SparkConf

# Initialize Spark Context
conf = SparkConf().setAppName("Sum using Accumulator").setMaster("local[*]")
sc = SparkContext(conf=conf)
# Create an RDD of a set of numbers
numbers_rdd = sc.parallelize([1, 2, 3, 4, 5])
# Initialize an accumulator with an initial value of 0
total_sum = sc.accumulator(0)
# Perform the sum using the accumulator
numbers_rdd.foreach(lambda x: total_sum.add(x))
# Get the result
print(f"Sum of numbers: {total_sum.value}")
```
